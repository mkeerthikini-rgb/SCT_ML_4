```python
import os
import zipfile
import cv2
import numpy as np

from google.colab import files
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

uploaded = files.upload()
zip_file = list(uploaded.keys())[0]

print("Uploaded ZIP File:", zip_file)

with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall('/content/gesture_data')

print("ZIP file extracted successfully!")

# Determine the actual directory containing images
base_extract_path = '/content/gesture_data'
extracted_items = os.listdir(base_extract_path)

image_directory = base_extract_path

# If there's only one item and it's a directory, assume it's the actual image folder
if len(extracted_items) == 1 and os.path.isdir(os.path.join(base_extract_path, extracted_items[0])):
    image_directory = os.path.join(base_extract_path, extracted_items[0])

# Now list images from the correct directory
images_found = os.listdir(image_directory)

print(f"Images Found in '{image_directory}':")
print(images_found)

data = []
labels = []

for image_filename in images_found:
    # Only process files that are images
    if image_filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        img_path = os.path.join(image_directory, image_filename)

        img = cv2.imread(img_path)

        if img is None:
            print(f"Skipping: {image_filename} (could not read image or it's not a valid image file)")
            continue
        img = cv2.resize(img, (64, 64))

        img = img.flatten()

        # Assuming image names contain 'thumb' or 'palm' to determine labels
        if 'thumb' in image_filename.lower():
            data.append(img)
            labels.append(0)
        elif 'palm' in image_filename.lower() or 'plam' in image_filename.lower(): # Corrected to include 'plam'
            data.append(img)
            labels.append(1)
        else:
            print(f"Warning: Image '{image_filename}' does not contain 'thumb' or 'palm'/'plam' in its name. Skipping.")

X = np.array(data)
y = np.array(labels)

# Check if X and y are empty before splitting
if X.size == 0 or y.size == 0:
    print("Error: No valid images with recognized labels were processed. Cannot perform train_test_split.")
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )

    model = SVC()

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)

    print("Model Trained Successfully!")
    print("Accuracy:", accuracy)

```
